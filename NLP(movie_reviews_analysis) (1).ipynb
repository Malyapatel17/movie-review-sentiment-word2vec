```python
#Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from tqdm import tqdm
corpus = []
for i in tqdm(range(0, len(messages))):
    view = re.sub('[^a-zA-Z]', ' ', messages['review'][i])
    view = view.lower()
    view = view.split()
    
    view = [lemmatizer.lemmatize(word) for word in view if not word in stopwords.words('english')]
    view = ' '.join(view)
    corpus.append(view)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess# It converts words to loercase()
```

```python
corpus[0]
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression  # or your model
import gensim
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Setup (run once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Load your trained Word2Vec model
model = gensim.models.Word2Vec.load("model_word2vec.model")  # or use pretrained

# Load your trained classifier (Random Forest / Logistic Regression)
import joblib
classifier = joblib.load("classifier_model.pkl")  # or use the trained one directly

# Lemmatizer
lemmatizer = WordNetLemmatizer()

# Average Word2Vec function
def avg_word2vec(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

# Predict function
def predict_sentiment(text):
    # 1. Preprocess
    tokens = word_tokenize(text.lower())
    lemmas = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha()]
    
    # 2. Vectorize
    vector = avg_word2vec(lemmas).reshape(1, -1)
    
    # 3. Predict
    prediction = classifier.predict(vector)[0]
    
    # 4. Map to label
    return "Positive 😊" if prediction == 1 else "Negative 😞"

# 🔍 Try it!
test_text = input("Enter your view:")
print(predict_sentiment(test_text))
```
